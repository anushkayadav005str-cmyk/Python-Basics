```python
# expression exeution rule1 = strings & numeric value can operate together with  this "*"
# rule2 = string &string can operate with '+'
#rule3= numeric value can operate with all arithmetic operators
# rule4 = arithmetic expressions with the integer and float will result in float

```

```python
# rule1=strings & numeric value can operate together with  this "*"
a,b=2,3
txt="@"
print(2*txt*3 )
print(type(txt))
```

```python
 # rule2=string &string can operate with '+'
a,b = '2',3
txt = "@"
print((a + txt )*b)
```

```python
# rule3= numeric value can operate with all arithmetic operaters
a,b,c=2,3,4
print(a+b*c)
```

```python
#rule4= arithmetic expressions with the integer and float will result float
a,b= 10,5.0
total = a*b
print(total)
```
